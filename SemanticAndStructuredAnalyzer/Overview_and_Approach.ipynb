# Prominent differences between the True News and Fake News Text:

 * ## Punctuation Errors:
       We can encounter various punctuation errors in fake news such as:
         - "don't" appear as "don t"
         - "he's" appear as "he s"
         - comma and other puntuation marks are also not used at proper place.
   
 * ## Spacing between words:
       Somtime there is no space between two words in a text and there are some instances where there are more than 1 spaces.
   
 * ## Presence of soical media handle in text:
       It can be seen that soical media handles like "(@realDonaldTrump)" can be seen in fake news but such articles are not present in true news.

 * ## Presence of urls in text:
       Harmful Url's such has "https://t.co/zcbyc4Wp5b" are present in fake news text.
   
 * ## Presence of tags at end of text:
       Tags such as "Photo by Spencer Platt/Getty Images.","Photo by Win McNamee/Getty Images.","Featured image via Justin Sullivan/Getty Images" can be seen at end of fake news.
   
 * ## Presence of tags at start of text:
       Tags such as "WASHINGTON (Reuters)" of authenticate news sources are present at start of text which are absent in fake news.
   
 * ## Repetiton of some specific type of words across different text articles:
       From the analysis of fake news text it was found that specific kind of words of type "2017{word}" was found across various fake news articles.
       Example: "2017he","2017ROFL","2017You"
   
 * ## Presence of url's of certain twitter posts:
       Url's of twitter post were seen to be found in the fake news articles.
       Example : "pic.twitter.com/XtZW5PdU2b","pic.twitter.com/fWfXsZupxy"
   
 * ## Presence of Hashtags:
       Hashtags were seen in fake news articles.
       Example: "#MeToo"
   
 * ## Absence of space after sentence completion:
       In some of the fake news it was seen that space was not given after the sentence completion while it was rarely seen in true news.
       Example: "international affairs.While the fact that most "
   
 * ## Absence of capitalizing the first letter after full stop:
       Whether the first letter of word after the full stop is capitalized or not.
       It helps to understand whether the writer has knowledge about the writing styles or not.
       This provides information whether the news writer is the professional one or some random on has written the news.
   
 * ## Capitalizing less significant words in article:
       Less significant words were entirely capitalized in some fake news articles.
       Example: "ZERO","PROBABLY","VERY"
   
 * ## Words with wrong spelling in the text:
       Again this help to get the idea about the writing knowledge of the writer of the text article.
   
   
 * ## Presence of Hate Speech:
       Some instances of hate speech were seen in fake news articles.
       Example: "President Liar Pants","President Angry Pants"
   
 * ## Repetion of some particular words throughout the text:
       It was seen that certain specific words were repeated throughout the fake news article which don't make any sense.
       Example: "Pew Research Center" and "fake news media" were repeated in the samples 1 and 5 of "Sample Text of Fake News" respectively
   
   

# What cannot be done:
 * ## Converting the letters to lowercase :
        As it is seen the Uppercase letters randomly present within the text can be a pattern to detect fake news.
   
 * ## Removing the punctuations from text :
        It is one of the pattern to detect the fake news.
   
 * ## Giving the entire 'text' column to our model:
        The average num_of_words in text field of both true news and fake news is 367 and 373 respectively.
        Huge context window required.Not Possible
   
 * ## Just providing the 'title' column to our model:
        The title feature would not be enough to accurately predict whether the news is fake or not.

# Approach:

## Approach1 : Machine Learning Model:

  ### Feature Engineering:

   The first step we have to perform before we trained Machine Learning model is Feature Engineering.
   We have to create features for our model.
   Given are the features which can be created from the text data provided:

  ####  1) Number of single quotes error in text
  ####  2) Number of Spacing error in text
  ####  3) Number of social media handle in text
  ####  4) Number of url's present in text
  ####  5) Presence of author/source tag at the end of text (Yes/No)
  ####  6) Presence of author/source tag at the start of the text (Yes/No)
  ####  7) Number of url's of twitter post
  ####  8) Number of hastags in text
  ####  9) Number of times absence of space after completion of sentence
  #### 10) Number of entirely capatalized word in text.
  #### 11) Number of Hate Speech instances present:
    Here we have to give the text to llm and asked the llm to return instances that depict hate speech.

  #### 12) Number of some common patterns of word that are present across different articles:
    Here first we have to find those types of word which are present across different articles.
    For example the words of type "2017{word}" are found across different fake news articles.
    So then one of our feature would be "Numbers of words of type 2017{word} in text".
    
 ####  13) Number of words with wrong spelling.
 

  ### Important Note:
    We will not use the feature "Presence of author/source tag at the end of text (Yes/No)" and "Presence of author/source tag at the start of the text (Yes/No)" because every true news article has source tag at start and every fake news article has a tag of 'Feature image by' at the end of the article.This features will just act as a noise and the model won't be generalize well.
    
  So these are the features which can then be feed to our model to get the prediction whether the news is Fake(0) or True(1).
  It is a classification problem

## Approach2: Network of LSTM's:
* We will use the layers of LSTM model to analyze the semantic meaning,grammactical structutre of the model.
* The "text" column will be the input to these models.
* But the mean number of words in a "text" field is approximately 360 words.And we cannot provide this large context window to our model.
* So we have to divide the "text" field.And pass every divided text to different lstm networks.And at last combine the results to produce the output.
