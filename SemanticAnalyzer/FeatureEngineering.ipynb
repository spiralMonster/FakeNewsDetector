## FEATURE - SINGLE QUOTE ERROR

```python
def check_single_quote_error(text:str):
    num_errors=0
    list_words=text.split(" ")
    ind1=0
    
    while ind1<len(list_words)-1:
        ind2=ind1+1
        word1=list_words[ind1]
        word2=list_words[ind2]
        
        if word1.isalpha() and word2.isalpha():
            word1=word1.lower()
            word2=word2.lower()
            
            if len(word1)>1 and len(word2)==1:
                if(word2) not in ['a','i']:
                    num_errors+=1
                    
            elif word1=="i" and word2=="ll":
                num_errors+=1
            
        ind1+=1
        
    return num_errors
    
        
```

## FEATURE - SPACING ERROR

```python
def check_spacing_error(text:str):
    num_errors=0
    list_words=text.split(" ")
    
    for word in list_words:
        if word=="":
            num_errors+=1
            
    return num_errors
    
        
```

## FEATURE - SOCIAL MEDIA HANDLE IN TEXT

```python
def check_social_media_handle(text:str):
    num_handles=0
    list_words=text.split(" ")
    
    for word in list_words:
        if "@" in word:
            num_handles+=1
            
    return num_handles
```

## FEATURE - SPACE ABSENCE AFTER SENTENCE COMPLETION

```python
def check_space_absence_after_sentence_completion(text:str):
    num_errors=0
    list_words=text.split(" ")
    
    for word in list_words:
        if "." in word:
            if "."!=word[-1]:
                num_errors+=1
                
    return num_errors
                
        
```

## FEATURE - CAPATALIZED WORDS

```python
import string
def check_capatalized_words(text:str):
    table=str.maketrans("","",string.punctuation)
    num_capitalized_words=0
    list_words=text.split(" ")
    
    for word in list_words:
        word=word.translate(table)
        if word.isalpha():
            if word.isupper():
                if word!="I":
                    num_capitalized_words+=1
                
    return num_capitalized_words
```

## FEATURE - Capitalizing Word After Full Stop

```python
def check_capitalization_after_full_stop(text:str):
    num_errors=0
    list_sent=text.split(".")
    list_sent=[sent for sent in list_sent if sent!='']
    list_sent=[sent.strip() for sent in list_sent]
    
    for sent in list_sent:
        if sent!="":
            first_letter=sent[0]
            if first_letter.isalpha():
                if not first_letter.isupper():
                    num_errors+=1
                
    return num_errors
```

## FEATUTE - SPELLCHECKER

```python
from spellchecker import SpellChecker
import string
```

```python
spell=SpellChecker()
table=str.maketrans('','',string.punctuation)
def check_spelling(text:str):
    num_errors=0
    list_words=text.split(" ")
    
    for word in list_words:
        if "." not in word:
            word=word.translate(table)
            if word.isalpha():
                word=word.lower()
                if word not in spell:
                    num_errors+=1
                
    return  num_errors
```
