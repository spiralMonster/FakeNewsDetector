```python
import matplotlib.pyplot as plt
```

```python
def plot_histograms(dataframe_true_news,dataframe_fake_news,column):
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1,2,1)
    max_val=dataframe_true_news[column].max()
    plt.hist(dataframe_true_news[column],color="blue")
    plt.xlabel(column)
    plt.ylabel("Frequency")
    plt.xlim(0,max_val)
    plt.title(f"{column} in true news articles")
    
    plt.subplot(1,2,2)
    max_val=dataframe_fake_news[column].max()
    plt.hist(dataframe_fake_news[column],color="orange")
    plt.xlabel(column)
    plt.ylabel("Frequency")
    plt.xlim(0,max_val)
    plt.title(f"{column} in Fake news articles")
    
    plt.subplots_adjust(wspace=0.5)
    plt.show()
    
    
```

```python
for column in final_data.columns:
    if column!="label":
        plot_histograms(true_news_data,fake_news_data,column)
```

## KEY INSIGHTS:
* #### 50% of true news articles don't show any single quote errors.Whereas the average single quote error shown in 25,50 and 75 percent of the data are 3,6,10 respectively.So single quote error can be a deciding feature.
  
* #### For spacing error it is evident that on average 10-12 errors are seen in both true news data and fake news data. But the maximum spacing error in true news article was somewhat 3729 which is far higher than the maximum spacing error of fake news article which is 301.So we have to remove the outliers from the true news articles and also from the fake news articles for that we have to decide the threshold which will be done after further analysis.
  
* #### Social media handles are rarely seen in both true news articles and false news articles.In fact 75% of true news articles don't have any social media handle in them and on the contrary the fake news articles has 1 news articles on average in 75% of data. Some fake news articles show the social media handles in them.So from true news articles we have to remove those entries which are having a lot of social media handles.
  
* #### If we consider the features of 'Url in text','Hastag','Twitter posts url' then not a single true news article has this feature. And in fact 75% of fake news articles also don't have these feature. So these feature will just increase the noise in the data.We are dropping these features.
  
* #### Spacing Error after sentence completion can be a differentiating factor between true news article and fake news articles but still we have to eliminate the outliers.
  
* #### The distribution of number of capitalized words in text is almost identical between the true news article and fake news article.So this feature cannot be a deciding factor. So we will drop this one.

* #### The distribution of feature "Number of absence of capitalization after sentence completion" is almost same for 75% of data but there is possibility that for the rest 25 % of data there can be change in distribution.It is beacuse for true news article the maximum value was 29 whereas for fake news article it was 479.
  
* #### In case of words of pattern "2017{word}" are not found in 75% of data of both true news articles and false news articles but max values which are 4,28 for true news articles and false news articles respectively depict that they can be present in the rest 25% of data.
 
* #### For "Number of mispelled words" the distibution of data is same upto 75% of data for true news articles and false news articles but the distribution might change for next 25% of data.
